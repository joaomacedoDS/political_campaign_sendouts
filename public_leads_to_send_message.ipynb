```python
#função que adiciona 55 nos números com menos de 12 dígitos
def format_phone(phone):
    phone_str = str(phone)
    if len(phone_str) < 12:
        return '55' + phone_str
    return phone_str

# Aplicar a função na coluna 'phone'
df['phone'] = df['phone'].apply(format_phone)
```
