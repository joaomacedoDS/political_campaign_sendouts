```python
import requests
import concurrent.futures
import pandas as pd

url = "https://evolutionparadisparos.7bot.ai/message/sendText/roberto"

headers = {
    "apikey": "FA5B26AD7541-4488-8648-AB46E113E738",
    "Content-Type": "application/json"
}

# Fun√ß√£o para enviar mensagem variando contato e contar as respostas
def enviar_mensagem(contato):
    payload = {
        "text": "üö®üö®üö®üö®üö®üö®üö®üö®üö®üö®üö®üö®üö® \n\n Ol√° sou o Dudu!! \n\n Sou morador da Mar√© e vejo o Roberto Est√°cio, nosso Beto como o √∫nico candidato para receber nosso voto.‚úÖ \n\n O cara √© da Mar√©, defende a Mar√©, e n√£o corre de problema. \n\n N√£o vamos dar nossos votos a candidatos meia boca, aproveitadores que s√≥ aparecem em ano de elei√ß√£o, Vamos fazer a diferen√ßa dessa vez! üòÉ E colocar algu√©m que fale a nossa l√≠ngua e que conhe√ßa as dores de nosso dia a dia. \n\n\n  Pe√ßo seu voto e de sua fam√≠lia para elegermos o Roberto Est√°cio. \n\n Desde j√° agrade√ßo e que Deus te aben√ßoe sempre!üôèüôè",
        "delay": 1200,
        "number": contato
    }

    try:
        response = requests.post(url, json=payload, headers=headers)
        if response.status_code in [200, 201]:
            print(f"Mensagem enviada para ({contato}): {response.status_code}")
            return 1
        else:
            print(f"Falha ao enviar para ({contato}): {response.status_code}")
            return 0
    except Exception as e:
        print(f"Erro ao enviar para ({contato}): {e}")
        return 0

# Fun√ß√£o para enviar mensagens em massa variando contato
def enviar_mensagens_para_todos_contando(contatos):
    with concurrent.futures.ThreadPoolExecutor(max_workers=1000) as executor:
        results = list(executor.map(enviar_mensagem, contatos))
    return sum(results)
```
